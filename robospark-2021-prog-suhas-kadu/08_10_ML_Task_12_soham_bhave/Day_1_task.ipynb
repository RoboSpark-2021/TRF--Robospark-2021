** Hard: How many people have a credit card that expires in 2025? **

````python
result = df['CC Exp Date']
L = []
for i in result:
  res = i[i.index('/')+1:]
  L.append(res)
df1 = pd.Series(L)
L1 = []
for i in df1:
  if i == '25':
    L1.append(i)
print("How many people have a credit card that expires in 2025? ",len(L1))
````
