```python
import pandas as pd     
import numpy as np 
df = pd.read_csv(path)
df
```

```python
df.isnull().sum()
```

```python
df.drop(labels = {'cabin' ,'boat' , 'body' , 'home.dest' } , axis = 1 , inplace = True)
df
```

```python
col_null = ['age'] 
med = []
 
for i in col_null: 
  med.append(df[i].median())  
k = 0 

for i in col_null:
  df[i].fillna(med[k] , inplace = True) 
  k = k + 1 
```

```python
df=df.dropna()
df=df.drop(columns=["name"])
df
```

```python
df.isnull().sum()
```

```python
ls=[]
for i in df['ticket']:
  s=''
  for j in i:
    if j.isnumeric():
      s+=j
  if len(s)!=0:
    ls.append(int(s, 10))
  else:
    ls.append(0)
df['ticket']=ls
df
```

```python
print(df['embarked'].unique())
print(df['sex'].unique())
```

```python
from sklearn import preprocessing
label1 = preprocessing.LabelEncoder()

df['embarked']= label1.fit_transform(df['embarked'])
df['sex']= label1.fit_transform(df['sex']) 
```

```python
print(df['embarked'].unique())
print(df['sex'].unique())
df
```

```python
from sklearn import model_selection
X = df.drop('survived' , axis = 1) 
Y = df['survived']
X_train , X_test , y_train, y_test = model_selection.train_test_split(X , Y ,test_size = 0.3 , shuffle = True , stratify=Y  )
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```
